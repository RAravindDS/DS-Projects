# Model Building

```python
# This is the regression problem so we choose all the regression algorithms. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
r = []
mse = []
mae = []
rmae = []
def findbaby(ml_model):
    model = ml_model.fit(X_train, y_train)
    print("Training_Score: ", model.score(X_train, y_train))
    y_predictions = model.predict(X_test)
    print('Predictions are: ', y_predictions)
    r2 = r2_score(y_test, y_predictions)
    print('R2_Score:{}'.format(r2))
    print('-----😎😋😋😋😊😎😋-----')
    print('Mean_Square_Error', mean_squared_error(y_test, y_predictions))
    print('----------💞💤💞-------------')
    print('Mean_Absolute_Error', mean_absolute_error(y_test, y_predictions))
    print('--------------🎇🎆🧨----------------------')
    print('Root_Mean_Squarred_Error', np.sqrt(mean_absolute_error(y_test, y_predictions)))
    sns.displot(y_test-y_predictions,color = 'red', lw =5, fill = True)
    r.append(r2_score(y_test, y_predictions))
    mse.append(mean_squared_error(y_test, y_predictions))
    mae.append(mean_absolute_error(y_test, y_predictions))
    rmae.append(np.sqrt(mean_absolute_error(y_test, y_predictions)))
    
    
    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
findbaby(LinearRegression())
```

```python
findbaby(KNeighborsRegressor())
```

```python
findbaby(RandomForestRegressor())
```

```python
findbaby(DecisionTreeRegressor())
```

```python
from xgboost import XGBRegressor
```

```python
findbaby(XGBRegressor())
```

```python
e = {'r2_score': r, 'MSE': mse, 'MAE': mae, 'RMAE': rmae}
info_s = pd.DataFrame(e, index = ['LinearRegression', 'KNeighborsRegressor','RandomForestRegressor',
                               'DecisionTreeRegressor', 'XGBRegressor'])
```
